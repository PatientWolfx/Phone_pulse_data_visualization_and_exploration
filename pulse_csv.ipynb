```python
import pandas as pd
import json
import os

path1 = "C:\\Users\\abuba\\Streamlit\\Projects\\Phonepe_pulse\\Clone\\pulse\\data\\aggregated\\transaction\\country\\india\\state\\"
Agg_trans_list = os.listdir(path1)
#Agg_state_list

clm1 = {'State':[], 'Year':[], 'Quater':[], 'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}
for i in Agg_trans_list:
    p_i = path1+i+"\\"
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = p_i+j+"\\"
        Agg_yr_list = os.listdir(p_j)
        
        for k in Agg_yr_list:
            p_k = p_j+k
            #To load json file
            Data = open(p_k, 'r')
            D = json.load(Data)
            
            #To assign datas into variables
            for z in D['data']['transactionData']:
                Name = z['name']
                count = z['paymentInstruments'][0]['count']
                amount = z['paymentInstruments'][0]['amount']
                clm1['Transaction_type'].append(Name)
                clm1['Transaction_count'].append(count)
                clm1['Transaction_amount'].append(amount)
                clm1['State'].append(i)
                clm1['Year'].append(j)
                clm1['Quater'].append(int(k.strip('.json')))
                
Agg_trans = pd.DataFrame(clm1)
Agg_trans.to_csv('aggregated_transaction.csv', index = False)
```

```python
#for aggregated userwise_transactions 
path2 = "C:\\Users\\abuba\\Streamlit\\Projects\\Phonepe_pulse\\Clone\\pulse\\data\\aggregated\\user\\country\\india\\state\\"
Agg_user_list = os.listdir(path2)
clm2 = {'State':[], 'Year':[], 'Quater':[], 'Brand':[], 'Count':[], 'Percentage':[]}

for i in Agg_user_list:
    p_i = path2+i+'\\'
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = p_i+j+'\\'
        Agg_yr_list = os.listdir(p_j)
        for k in Agg_yr_list:
            p_k = p_j+k
            Data = open(p_k, 'r')
            load_data = json.load(Data)
            try:
                for z in load_data['data']['usersByDevice']:
                    clm2['Brand'].append(z['brand'])
                    clm2['Count'].append(z['count'])
                    clm2['Percentage'].append(z['percentage'])
                    clm2['State'].append(i)
                    clm2['Year'].append(j)
                    clm2['Quater'].append(int(k.strip('.json')))
            except:
                pass
Agg_user = pd.DataFrame(clm2)
Agg_user.to_csv('aggregated_user.csv', index = False)
```

```python
path3 = "C:\\Users\\abuba\\Streamlit\\Projects\\Phonepe_pulse\\Clone\\pulse\\data\\map\\transaction\\hover\\country\\india\\state\\"
Map_trans_list = os.listdir(path3)
clm3 = {'State':[], 'Year':[], 'Quater':[], 'District':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Map_trans_list:
    p_i = path3+i+'\\'
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = p_i+j+'\\'
        Agg_yr_list = os.listdir(p_j)
        for k in Agg_yr_list:
            p_k = p_j+k
            Data = open(p_k, 'r')
            load_data = json.load(Data)
            for z in load_data['data']['hoverDataList']:
                clm3['State'].append(i)
                clm3['Year'].append(j)
                clm3['Quater'].append(int(k.strip('.json')))
                District = z['name']
                count = z['metric'][0]['count']
                amount = z['metric'][0]['amount']
                clm3['District'].append(District)
                clm3['Transaction_count'].append(count)
                clm3['Transaction_amount'].append(amount)
Map_trans = pd.DataFrame(clm3)
Map_trans.to_csv('Map_transaction.csv', index = False)
```

```python
path4 = "C:\\Users\\abuba\\Streamlit\\Projects\\Phonepe_pulse\\Clone\\pulse\\data\\map\\user\\hover\\country\\india\\state\\"
Map_user_list = os.listdir(path4)
clm4 = {'State':[], 'Year':[], 'Quater':[], 'District':[], 'Users':[]}

for i in Map_user_list:
    p_i = path4+i+'\\'
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = p_i+j+'\\'
        Agg_yr_list = os.listdir(p_j)
        for k in Agg_yr_list:
            p_k = p_j+k
            Data = open(p_k, 'r')
            load_data = json.load(Data)
            for z in load_data['data']['hoverData'].items():#to loop through both keys and values
                clm4['State'].append(i)
                clm4['Year'].append(j)
                clm4['Quater'].append(int(k.strip('.json')))
                name = z[0]
                clm4['District'].append(name)
                user = z[1]['registeredUsers']
                clm4['Users'].append(user)

Map_user = pd.DataFrame(clm4)
Map_user.to_csv('Map_users.csv', index = False)
```
